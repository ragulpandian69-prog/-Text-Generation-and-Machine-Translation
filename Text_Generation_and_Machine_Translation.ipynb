Question 1: What is Generative AI and its primary use cases?

Answer:

Generative AI is a type of artificial intelligence that can create new content such as text, images, audio, and videos by learning patterns from existing data. It uses deep learning models like Generative Adversarial Networks (GANs), Variational Autoencoders (VAEs), and Transformer-based models.

Generative AI is widely used across industries. In healthcare, it helps in drug discovery and medical imaging. In finance, it is used for fraud detection and automated report generation. In marketing, it generates advertisements, social media content, and personalized campaigns. In entertainment, it is used to create stories, music, and videos. In e-commerce, it helps generate product descriptions and powers chatbots.

Question 2: Role of probabilistic modeling and difference from discriminative models

Answer:

Probabilistic modeling plays an important role in generative models because it allows the model to learn the underlying probability distribution of the data. Generative models try to learn the joint probability of input and output, which helps them generate new data that is similar to the training dataset.

On the other hand, discriminative models focus only on learning the relationship between input and output. They learn the conditional probability and are mainly used for classification tasks.

The key difference is that generative models can both understand and generate data, while discriminative models are mainly used to predict labels based on input data.

Question 3: Difference between Autoencoders and VAEs

Answer:

Autoencoders are neural networks used to compress input data into a smaller representation and then reconstruct it back. They work in a deterministic way, meaning the same input will always produce the same output.

Variational Autoencoders (VAEs) are an advanced version of autoencoders that learn the probability distribution of the data instead of fixed representations. They introduce randomness by learning mean and variance, which allows them to generate new data samples.

In simple terms, autoencoders are mainly used for reconstruction, while VAEs are used for generating new data similar to the original dataset.

Question 4: Attention mechanism in Neural Machine Translation

Answer:

Attention mechanism helps the model focus on important words in a sentence while translating. In Neural Machine Translation, the encoder processes the input sentence and the decoder generates the translated sentence one word at a time.

During this process, attention assigns different weights to different words in the input sentence based on their importance. This helps the model focus on relevant words instead of relying only on the last hidden state.

Attention is very important because it improves translation accuracy, especially for long sentences, and reduces information loss during translation.

Question 5: Ethical considerations in Generative AI

Answer:

There are several ethical issues to consider when using generative AI. One major concern is bias, as models may learn and reproduce biased patterns from training data. Another issue is plagiarism, where generated content may be very similar to existing work. Generative AI can also be misused to create fake content or misinformation, such as deepfakes. Copyright issues are also important, as it is not always clear who owns the generated content. Therefore, responsible use of generative AI is very important.

Question 6: VAE for text reconstruction

Answer:

In this task, a small dataset of sentences is first preprocessed by converting text into numerical form using tokenization. Then, sequences are padded so that all sentences have the same length. A Variational Autoencoder model is built using an encoder to convert input into latent variables (mean and variance) and a decoder to reconstruct the original sentences. The model is trained on the dataset and learns to reconstruct similar sentences

```python
#Question 6: VAE for text reconstruction

import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import layers, Model

texts = ["The sky is blue", "The sun is bright",
         "The grass is green", "The night is dark",
         "The stars are shining"]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

max_len = max(len(seq) for seq in sequences)
padded = pad_sequences(sequences, maxlen=max_len, padding='post')

vocab_size = len(tokenizer.word_index) + 1

inputs = layers.Input(shape=(max_len,))
embedding = layers.Embedding(vocab_size, 16)(inputs)
flat = layers.Flatten()(embedding)

z_mean = layers.Dense(8)(flat)
z_log_var = layers.Dense(8)(flat)

def sampling(args):
    z_mean, z_log_var = args
    # Keras expects symbolic tensors, not numpy arrays, for operations within the graph.
    # Using tf.random.normal instead of np.random.normal
    epsilon = tf.random.normal(shape=tf.shape(z_mean))
    return z_mean + tf.exp(0.5 * z_log_var) * epsilon

z = layers.Lambda(sampling, output_shape=(8,))([z_mean, z_log_var])

decoder = layers.Dense(max_len * vocab_size, activation='softmax')(z)
outputs = layers.Reshape((max_len, vocab_size))(decoder)

vae = Model(inputs, outputs)
vae.compile(optimizer='adam', loss='sparse_categorical_crossentropy')

vae.fit(padded, padded, epochs=10)
```

Question 7: GPT Translation

Answer:

A pre-trained model can translate text into different languages.

```python
#CODE QUESTION 7

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("t5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("t5-small")

text = "Artificial Intelligence is transforming the world."

# Translate to French
input_text_fr = "translate English to French: " + text
input_ids_fr = tokenizer(input_text_fr, return_tensors="pt").input_ids
outputs_fr = model.generate(input_ids_fr)
fr_translation = tokenizer.decode(outputs_fr[0], skip_special_tokens=True)

# Translate to German
input_text_de = "translate English to German: " + text
input_ids_de = tokenizer(input_text_de, return_tensors="pt").input_ids
outputs_de = model.generate(input_ids_de)
de_translation = tokenizer.decode(outputs_de[0], skip_special_tokens=True)

print(f"French: {fr_translation}")
print(f"German: {de_translation}")
```
